```python
import numpy as np
import pandas as pd
```

```python
df_csv = pd.read_csv('sales.csv')
df_csv.head()
```

```python
df_json = pd.read_json('sales.json')
df_json.head()
```

```python
df_excel = pd.read_excel('sales.xlsx')
df_excel.head()
```

```python
df_csv.isnull().sum()
```

```python
df_csv.dropna(inplace=True)
df_json.dropna(inplace=True)
df_excel.dropna(inplace=True)
```

```python
print(df_csv.shape)
print(df_json.shape)
print(df_excel.shape)
```

```python
df1 = df_csv.iloc[:9000, :]
df2 = df_json.iloc[:9000, :]
df3 = df_excel.iloc[:9000, :]
```

```python
print(df1.shape)
print(df2.shape)
print(df3.shape)
```

```python
df1 = df1.reset_index(drop=True)
df2 = df2.reset_index(drop=True)
df3 = df3.reset_index(drop=True)

df_merged = pd.concat([df1, df2, df3], axis=1)
```

```python
df_merged.dropna(inplace=True)
```

```python
df_merged
```

```python
df_merged.columns
```

### 1. Bar Plot - Sales by Product Category:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named df
df_merged.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.show()
```

### 2. Pie Chart - Sales Distribution by Region:

```python
df_merged.groupby('Region')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales Distribution by Region')
plt.show()
```

### 3. Box Plot - Sales Distribution by Product Category:

```python
import seaborn as sns

sns.boxplot(x='Category', y='Sales', data=df_merged)
plt.title('Sales Distribution by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Sales')
plt.show()
```
